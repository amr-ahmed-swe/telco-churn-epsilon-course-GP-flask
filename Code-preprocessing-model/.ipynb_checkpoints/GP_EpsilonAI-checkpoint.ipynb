```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 40, stratify=y)
```

```python
colum = list(X.columns.values)
colum
```

```python
scaler= StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

```

```python
classifiers = {}
```

```python
knn_model = KNeighborsClassifier(n_neighbors = 9) 
knn_model.fit(X_train,y_train)
predicted_knn = knn_model.predict(X_test)
accuracy_knn = metrics.accuracy_score(y_test, predicted_knn)
print("KNN accuracy:",accuracy_knn)
classifiers['KNNClassifier'] = accuracy_knn
```

```python
svc_model = SVC(random_state = 1)
svc_model.fit(X_train,y_train)
predict_svc = svc_model.predict(X_test)
accuracy_svc = metrics.accuracy_score(y_test, predict_svc)
print("SVM accuracy is :",accuracy_svc)
classifiers['SVC'] = accuracy_svc
```

```python
model_rf = RandomForestClassifier(n_estimators=504 , oob_score = True, n_jobs = -1,
                                  random_state =50, max_features = "auto",
                                  max_leaf_nodes = 30)
model_rf.fit(X_train, y_train)

# Make predictions
prediction_test = model_rf.predict(X_test)
accuracy_RFC = metrics.accuracy_score(y_test, prediction_test)
print ('accuracyof Random Forest: ', accuracy_RFC)

classifiers['RandomForestClassifier'] = accuracy_RFC
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
prediction_lr = model_rf.predict(X_test)
accuracy_lr = metrics.accuracy_score(y_test, prediction_lr)
print("Logistic Regression accuracy is :",accuracy_lr)
classifiers['Logistic Regression'] = accuracy_lr
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
predictdt_DT = dt_model.predict(X_test)
accuracy_dt = metrics.accuracy_score(y_test, predictdt_DT)
print("Decision Tree accuracy is :",accuracy_dt)

classifiers['DecisionTreeClassifier'] = accuracy_dt
```

```python
a_model = AdaBoostClassifier(n_estimators=25,learning_rate=1)
a_model.fit(X_train,y_train)
ada_preds = a_model.predict(X_test)
ada_score = metrics.accuracy_score(y_test, ada_preds)
print("AdaBoost Classifier accuracy:", ada_score)
classifiers['AdaBoostClassifier'] =  ada_score
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
gb_score = metrics.accuracy_score(y_test, gb_pred)
print("Gradient Boosting Classifier", gb_score)
classifiers['GradientBoostingClassifier'] =  gb_score
```
